```python
#Imagem 1 - Função Gama - 3.9
def adjust_gamma(image, gamma=1.0):
    table = np.array([((i / 255.0) ** gamma) * 255
        for i in np.arange(0, 256)]).astype("uint8")
    return cv.LUT(image, table)
```

```python
gama = [1, 3, 4, 5]
img = imread("entradas/"+imagens[0])
imagem1 = adjust_gamma(img,gama[0])
imgplot = plt.imshow(imagem1, cmap='gray', vmin=0, vmax=255)
```

# <p id= 'imagem2'>Imagem 3.10</p>

```python
def alargamento_de_contraste(image):
  min = np.amin(image)
  max = np.amax(image)
  return ((image - min)/(max - min))*255
```

```python
def limiarizacao(image):
  min = np.mean(image)
  return (image - min)*255
```

```python
image = imread("entradas/"+imagens[1])
imshow(image, cmap='gray', vmin=0, vmax=255);
```

```python
image_applied = alargamento_de_contraste(image)
imshow(image_applied, cmap='gray', vmin=0, vmax=255);
```

```python
image_applied = limiarizacao(image)
imshow(image_applied, cmap='gray', vmin=0, vmax=255);
```

# <p id= 'imagem3'>Imagem 3.12</p>

```python
def fatiamento(image, A, B):
    image_applied = image.copy()
    image_applied[image < A] = 0
    image_applied[np.logical_and(image > A, image < B)] = 255
    return image_applied
```

```python
def fatiamento2(image, A, B):
    image_applied = image.copy()
    image_applied[np.logical_or(image < A, image > B)] = 30
    image_applied[image < 80] = 2*image_applied[image < 80]
    return image_applied
```

```python
image = imread('entradas/' + imagens[2])
imshow(image, cmap='gray', vmin=0, vmax=255);
```

```python
image_applied = fatiamento(image, 140, 255)
imshow(image_applied, cmap='gray', vmin=0, vmax=255);
```

```python
image_applied = fatiamento2(image, 146, 255)
imshow(image_applied, cmap='gray', vmin=0, vmax=255);
```

# <p id= 'imagem5'>Imagem 3.20</p>

```python
def count_intensities(image, num_pixels):
  image_one_array = np.reshape(image, num_pixels)
  intensity_array = np.zeros(256)
  for pixel in image_one_array:
      intensity_array[pixel] += 1

  return intensity_array
```

```python
def equalization_histogram(image, intensity, num_pixels):
  image_one_array = np.reshape(image, num_pixels)

  sum_acum = np.cumsum(intensity)

  n = (sum_acum - sum_acum.min()) * 255
  N = sum_acum.max() - sum_acum.min()
  c = n/N
  c = c.astype('uint8')
  
  return c[image_one_array] 
```

```python
image = imread('entradas/' + imagens[4])
imshow(image, cmap='gray', vmin=0, vmax=255);
```

```python
equa = cv.equalizeHist(src=image)
cv.calcHist(equa, [0], None, [256], [0, 256])
plt.hist(equa.ravel(), 256, [0, 256])
plt.title('Histograma Equalizado')
plt.xlabel('Valores dos pixels')
plt.ylabel('Qntd. de pixels')
plt.grid(True)
```

```python
imshow(equa, cmap='gray', vmin=0, vmax=255);
```
