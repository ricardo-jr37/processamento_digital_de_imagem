```python
#Filtro Passa-Baixa Ideal
def ILPF(D0,img):
    shape = img.shape
    imagem = np.zeros(shape[:2])
    rows, cols = shape[:2]
    center = (rows/2,cols/2)
    for x in range(cols):
        for y in range(rows):
            if math.dist((y,x),center) < D0:
                imagem[y,x] = 1
    return imagem
```

# <p id= 'imagem10'>Imagem 4.45</p>

```python
def BLPF(D0, img, n):
    shape = img.shape
    imagem = np.zeros(shape[:2])
    rows, cols = shape[:2]
    center = (rows/2, cols/2)
    #Percorrendo toda a imagem
    for x in range(cols):
        for y in range(rows):
            imagem[y,x] = 1/(1+(math.dist((y,x), center)/D0)**(2*n))
            
    return imagem
```

# <p id= 'imagem11'>Imagem 4.48</p>

```python
def GLPF(D0,img):
    shape = img.shape
    imagem = np.zeros(shape[:2])
    rows, cols = shape[:2]
    center = (rows/2,cols/2)
    
    for x in range(cols):
        for y in range(rows):
            imagem[y,x] = np.exp(((-math.dist((y,x),center)**2)/(2*(D0**2))))
    
    return imagem
```

# <p id= 'imagem13'>Imagem 4.55</p>

```python
def BHPF(D0,img,n):
    shape = img.shape
    imagem = np.zeros(shape[:2])
    rows, cols = shape[:2]
    center = (rows/2,cols/2)
    for x in range(cols):
        for y in range(rows):
            imagem[y,x] = 1-1/(1+(math.dist((y,x),center)/D0)**(2*n))
    
    return imagem
```

# <p id= 'imagem14'>Imagem 4.57</p>

```python
def limiarizacao(image):
    min = np.mean(image)
    return (image - min)*255
```

```python
def limit(img):
    shape = img.shape
    imagem = np.zeros(shape[:2])
    rows, cols = shape[:2]
    center = (rows/2,cols/2)
    min = np.mean(img)
    for x in range(cols):
        for y in range(rows):
                if img[y,x] > min:
                    imagem[y,x] = 255
                else:
                    imagem[y,x] = 0
    return imagem
```

```python
img = imread("entradas/"+imagens[9])
imshow(img, cmap='gray', vmin=0, vmax=255);
```

```python
fourier = np.fft.fft2(img)
fourier_centralizado = np.fft.fftshift(fourier)
fourier_centralizado_filtered = fourier_centralizado * ILPF(80, img)
fourier_filtered = np.fft.ifftshift(fourier_centralizado_filtered)
inversa = np.fft.ifft2(fourier_filtered)
plt.imshow(np.abs(inversa), "gray")
```

```python
limi = np.abs(inversa)
limi = limit(limi)
plt.imshow(limi, "gray")
```

```python
limi = np.abs(inversa)
limi = limiarizacao(limi)
plt.imshow(limi, "gray")
```
